## Resimdeki Yüzün Tespit Edilip Ayrı Bir Resim Olarak Kaydedilmesi

```python
import cv2
import numpy as np
```

```python
foto = '/path'

# Load the face cascade
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

# Read the input image
image = cv2.imread(foto)

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces in the image
faces = face_cascade.detectMultiScale(gray, 1.1, 10)
```

```python
#Detect the main picture on the card
area_list = []

for (x, y, w, h) in faces:
    area_list.append(w*h)

if len(area_list) > 0: #Resimde yüz tespiti yapılma durumu
    index = area_list.index(max(area_list))
else: #Resimde yüz tespiti yapılamazsa
    print("Yüz tespiti yapılamamıştır. Geçerli bir fotoğraf yükleyiniz.")
```

```python
x, y, w, h = faces[index]

# Draw a rectangle around face
cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Save the output image
cv2.imwrite(foto + '_output' + '.jpg', image)

face = image[y:y+h, x:x+w]  # Crop the face region
output_path = foto + '_face' + '.jpg' # Construct the output path for each face
cv2.imwrite(output_path, face)  # Save the cropped face as a separate image
```

```python
# Display the photo in Jupyter Notebook
from IPython.display import Image
display(Image(filename=foto, height=250, width=300))
display(Image(filename=foto + '_output' + '.jpg', height=250, width=300))
display(Image(filename=foto + '_face' + '.jpg', height=250, width=300))
```

## Kullanıcıdan Alınan Videodan Her Saniye Resim Çıkartılması

```python
# Videoyu aç
cap = cv2.VideoCapture("/path")

# Videonun süresini al
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Videodan saniye başı resim çıkarma
for i in range(0, frame_count, 24):
    cap.set(cv2.CAP_PROP_POS_FRAMES, i)
    ret, frame = cap.read()
    if ret == False:
        break

    # Resmi kaydet
    cv2.imwrite(f"image{i}.jpg", frame)

# Videoyu kapat
cap.release()
```
